```python
# Chargement de la donnée
# Ouverture du fichier en mode lecture ('r')
with open('data/raw/tiktok/user_data_tiktok.json', 'r', encoding='utf-8') as data:
    tiktok_data = json.load(data)
```

### Chargement des commentaires en DF

```python
#charging the comments List into a dataframe
commentsList = tiktok_data["Comment"]['Comments']['CommentsList']
df_comments = pd.DataFrame(commentsList)
print(df_comments.head())
```

```python
#deleting empty columns
clean_comments_df = df_comments.drop(columns=['photo', 'video', 'url'])
```

```python
print(clean_comments_df.head())
```

```python
#convert the date column to datetime
clean_comments_df['date'] = pd.to_datetime(clean_comments_df['date'])
```

```python
print(type(clean_comments_df['date'][0]))
```

```python
print(clean_comments_df.head())
```

### Chargement des publications(Post) en DF

```python
videoList = tiktok_data["Post"]['Posts']['VideoList']
df_videoList = pd.DataFrame(videoList)
print(df_videoList.head())
```

```python
# Suppression des colonnes inutiles
clean_videoList_df = df_videoList.drop(columns=['Link', 'AllowStitches', 'AllowDuets', 'AllowStickers', 'AllowSharingToStory', 'ContentDisclosure', 'AIGeneratedContent', 'Location', 'Title', 'AddYoursText', 'AlternateText', 'CoverImage', 'NumberOfCollections' ])
```

```python
print(clean_videoList_df.head())
```

```python
#convert the date column to datetime
clean_videoList_df['Date'] = pd.to_datetime(clean_videoList_df['Date'])
```

```python
print(type(clean_videoList_df['Likes'][0]))
```

```python
# Convert Likes into integer for future operations
clean_videoList_df['Likes'] = pd.to_numeric(clean_videoList_df['Likes'], errors='coerce')
print(type(clean_videoList_df['Likes'][0]))
```

```python
clean__watch_history_df = df_watch_history.drop(columns=['Link'])
print(clean__watch_history_df.head())
```

## Feature engineering

```python
print(clean_videoList_df.head())
```

```python
# Extraction du mois dans une nouvelle colonne sur clean_videoList_df
clean_videoList_df['Month'] = clean_videoList_df['Date'].dt.to_period('M')
```

```python
print(clean_videoList_df.head())
```

```python
clean_videoList_df['DayOfWeek'] = clean_videoList_df['Date'].dt.day_name()
print(clean_videoList_df.head())
```

```python
clean_videoList_df['Hour'] = clean_videoList_df['Date'].dt.hour
print(clean_videoList_df.head())
```

```python
clean_videoList_df['Quarter'] = clean_videoList_df['Date'].dt.quarter
print(clean_videoList_df.head())
```

```python
clean_videoList_df['SoundType'] = np.where(clean_videoList_df['Sound'].str.startswith("original sound"), 'Original', 'trending/music')
print(clean_videoList_df.head())
```
